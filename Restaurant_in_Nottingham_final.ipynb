```python
search_ff = 'Fast Food'
search_rs = 'Restaurant'
radius = 2000
```

```python
urlFF = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_ff, radius, LIMIT)
urlRT = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_rs, radius, LIMIT)
```

```python
resultsFF = requests.get(urlFF).json()
resultsRT = requests.get(urlRT).json()
#results
```

```python
#73 records obtained for the searched category
```

```python
# assign relevant part of JSON to venues
venues= resultsFF['response']['venues'] + resultsRT['response']['venues']
#print(venues)
# tranform venues into a dataframe
df1 = json_normalize(venues)
print('Size=',df1.shape)
df1.head()
```

```python
#Drop unnecessary columns from dataframe df1 and create a new dataframe df_new
```

```python
df2 = df1.drop(columns = ['hasPerk','id','location.cc','location.city','location.crossStreet','location.formattedAddress','location.labeledLatLngs','location.neighborhood','location.state','referralId','venuePage.id'])
df2.head(5)
```

```python
#Clean up the 'Categories' data string using Regular Expression to extract the 'name' field which has the cuisine type
```

•The Foursquare call doesn’t return the ‘cuisine type’ as a field in the JSON. Some of the fields such as pluralName or shortName fields have values which are used to extract the cuisine


```python
import re
i=0 #index to set clean values of 'Cuisine' for 'categories'
df3 = df2
for line1 in df3.categories:
    #print(df3.categories)
    name1 = str(re.findall(r"'name': '(.*?)'",str(line1)))
    df2.categories[i] = name1[2:-2]
    i=i+1
    #print('cuisine=',name1)
df2.head(10) 
```

```python
#Rename columns in the dataframe df_new to soemthing more meaningful
```

```python
df_clean = df2.rename(columns={'categories':'Cuisine','location.address':'Address','location.country' : 'Country','location.distance':'Distance(Mts.)','location.lat':'Latitude', 'location.lng':'Longitude','location.postalCode':'Postcode', 'name':'Name'})   
df_clean.head(5)
```

```python
#first find the number of rows with 'NaN' values
```

Some of the crucial fields such as Address, and Postcode return NaN data values. If the count is statistically insignificant, we can drop these rows. However, if they are too many we’ll need to deploy some other method to obtain the postcode and address. Useed geocoder to obtain the postcode and address via the Latitude and Longitude 
    • NaN is defined as a float, we need to identify those data values and decide whether to drop those rows, extrapolate the data or do something else.        Based on the importance and feasibility of extrapolating the data the following rule was adopted if a field contained unusable value, i.e. NaN 
    • Cuisine: Drop the row as without the cuisine there can be no recommendations made
    • Address: Keep it and check if it can be extrapolated from the Latitude, Longitude
    • Latitude/Longitude: Keep it and check if these can be extrapolated from the Address and Postcode
    • Postcode: Keep it and check if it can be extrapolated from Address, Latitude and Longitude


```python
Cuisine_NaN_count = 0
Address_NaN_count = 0
Latitude_NaN_count = 0
Longitude_NaN_count = 0
Postcode_NaN_count = 0
Name_NaN_count = 0

for ix in df_clean.index:
    if type(df_clean.Cuisine[ix]) is float:
        Cuisine_NaN_count = Cuisine_NaN_count+1
    if type(df_clean.Address[ix]) is float:
        Address_NaN_count = Address_NaN_count+1
    if type(df_clean.Latitude[ix]) is float:
        Latitude_NaN_count = Latitude_NaN_count+1
    if type(df_clean.Longitude[ix]) is float:
        Longitude_NaN_count = Longitude_NaN_count+1
    if type(df_clean.Postcode[ix]) is float:
        Postcode_NaN_count = Postcode_NaN_count+1
    if type(df_clean.Name[ix]) is float:
        Name_NaN_count = Name_NaN_count+1
print('Cuisine has junk value in ',Cuisine_NaN_count,' rows')
print('Address has junk value in ',Address_NaN_count,' rows')
print('Latitude has junk value in ',Latitude_NaN_count,' rows')
print('Longitude has junk value in ',Longitude_NaN_count,' rows')
print('Postcode has junk value in ',Postcode_NaN_count,' rows')
print('Name has junk value in ',Name_NaN_count,' rows')

Address_Postcode_Nan_count = 0
for iy in df_clean.index:
    if (type(df_clean.Address[iy]) is float) and (type(df_clean.Postcode[iy]) is float):
        Address_Postcode_Nan_count = Address_Postcode_Nan_count + 1
print('Both Address and Postcode has junk value in ',Address_Postcode_Nan_count,' rows')
```

```python
Obtain the missing  Address and postcode from Latitude and Longitude thru Geocoders API
```

**All the rows have Latitude and Longitude available. Use Geocoder to obtain the missing Postcode and Address

```python
from  geopy.geocoders import Nominatim
Ext_Address_count = 0
Ext_Postcode_count = 0
for iy in df_clean.index:
    if type(df_clean.Address[iy]) is float :
        geolocator = Nominatim(user_agent='my-application')
        latitude = str(df_clean.Latitude[iy])
        longitude = str(df_clean.Longitude[iy])
        loc = geolocator.geocode(latitude+','+ longitude)
        location = str(loc).split(',')
        df_clean.Address[iy] = location[0:3]
        Ext_Address_count = Ext_Address_count + 1
    if type(df_clean.Postcode[iy]) is float :
        geolocator = Nominatim(user_agent='my-application')
        latitude = str(df_clean.Latitude[iy])
        longitude = str(df_clean.Longitude[iy])
        loc = geolocator.geocode(latitude+','+ longitude)
        location = str(loc).split(',')
        df_clean.Postcode[iy] = location[-2]
        Ext_Postcode_count = Ext_Postcode_count + 1
print ('Obtained Address= ',Ext_Address_count,'Postcode= ',Ext_Postcode_count)
```
